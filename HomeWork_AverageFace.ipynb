```python
import numpy as np
import skimage 
from skimage import io
from matplotlib import pyplot as plt
myTensor=np.zeros((250,250,3),dtype=np.float64)
myTensor
```

```python
myIc=skimage.io.imread_collection("lfw/*/*.jpg")[:100]
```

```python
for image in myIc:
  myTensor+=image
myTensor/100
```

```python
skimage.io.imsave("MyNewImage.jpg",myTensor)
```

```python
skimage.io.imshow("MyNewImage.jpg")
```

```python
myNewImageToUint=skimage.io.imsave("MyNewImage2.jpg",(myTensor).astype(np.uint8))
```
